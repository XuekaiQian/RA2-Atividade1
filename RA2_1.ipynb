```python
#Alunos: Xuekai Qian, Rafael Leal, Marcelo Wzorek, Raphael de Oliveira
import numpy as np
import mpmath as mp
import matplotlib.pyplot as plt

# Define a função f(x)
def f(x):
  return 2*x + mp.sin(3*x)

# Define a derivada analítica de f(x)
def df_analytical(x):
  return 2 + 3*mp.cos(3*x)

# Define a fórmula das diferenças finitas para calcular a derivada de f(x)
def df_finite_difference(x, h):
  return (f(x+h) - f(x-h))/(2*h)

# Define a fórmula do ponto médio para calcular a derivada de f(x)
def df_midpoint(x, h):
  return (f(x + h/2) - f(x - h/2))/h

# Define o ponto onde a derivada é descendente
x_descendente = 1

# Define o ponto onde a tangente é zero
x_tangente_zero = mp.asin(-2/3)/3

# Define a precisão desejada
precisao = mp.mpf('1e-10')

# Calcula a derivada analítica de f(x) no ponto x_descendente
df_an = df_analytical(x_descendente)

# Calcula as derivadas numéricas de f(x) no ponto x_descendente
df_fd = df_finite_difference(x_descendente, precisao)
df_mp = df_midpoint(x_descendente, precisao)
# Calcula os erros absolutos das derivadas numéricas de f(x) no ponto x_descendente
abs_error_fd = mp.fabs(df_fd - df_an)
abs_error_mp = mp.fabs(df_mp - df_an)

# Calcula o erro absoluto entre as três derivadas de f(x) no ponto x_tangente_zero
df_an_tangente_zero = df_analytical(x_tangente_zero)
abs_error_fd_tangente_zero = mp.fabs(df_finite_difference(x_tangente_zero, precisao) - df_an_tangente_zero)
abs_error_mp_tangente_zero = mp.fabs(df_midpoint(x_tangente_zero, precisao) - df_an_tangente_zero)



# Imprime os resultados
print(f"Derivada analítica: {df_an}")
print(f"Derivada usando a Fórmula das Diferenças Finitas: {df_fd}")
print(f"Derivada usando a Fórmula do Ponto Médio: {df_mp}")
print(f"Erro absoluto Fórmula das Diferenças Finitas: {abs_error_fd}")
print(f"Erro absoluto Fórmula do Ponto Médio: {abs_error_mp}")
print(f"Erro absoluto entre as três derivadas possíveis usando a Fórmula das Diferenças Finitas: {abs_error_fd_tangente_zero}")
print(f"Erro absoluto entre as três derivadas possíveis usando a Fórmula do Ponto Médio: {abs_error_mp_tangente_zero}")

x_range = np.arange(0, 2, 0.01)
y_an = np.array([df_analytical(xx) for xx in x_range])
y_fd = np.array([df_finite_difference(xx, precisao) for xx in x_range])
y_mp = np.array([df_midpoint(xx, precisao) for xx in x_range])

plt.plot(x_range, y_an, label="Derivada analítica")
plt.plot(x_range, y_fd, label="Fórmula das Diferenças Finitas")
plt.plot(x_range, y_mp, label="Fórmula do Ponto Médio")
plt.legend()
plt.show()
```
