```python
#a
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 40*x**3 + 12*x**2 - 9*x
integral = sp.integrate(f, (x, 2.5, 7))

print("Valor da integral:", integral)

x_vals = np.linspace(2.5, 7, 100)
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]

plt.plot(x_vals, f_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função 40x^3 + 12x^2 - 9x')
plt.grid(True)
plt.show()
```

```python
#b
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

t = sp.symbols('t')
f = 12*t**7 - t**2 - t + 3
integral = sp.integrate(f, (t, 10, 7.5))

print("Valor da integral:", integral)

t_vals = np.linspace(7.5, 10, 100)
f_vals = [sp.N(f.subs(t, val)) for val in t_vals]

plt.plot(t_vals, f_vals)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.title('Gráfico da função 12t^7 - t^2 - t + 3')
plt.grid(True)
plt.show()
```

```python
#c
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 2*x**5 - x + 3*x**2
integral = sp.integrate(f, (x, 2, 1))

print("Valor da integral:", integral)

# Plotando o gráfico da função
x_vals = np.linspace(1, 2, 100)
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]

plt.plot(x_vals, f_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função 2x^5 - x + 3x^2')
plt.grid(True)
plt.show()
```

```python
#d
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 7*sp.sin(x) - 2*sp.cos(x)
integral = sp.integrate(f, (x, 2, 0))

print("Valor da integral:", integral)


x_vals = np.linspace(0, 2, 100)
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]

plt.plot(x_vals, f_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função 7*sin(x) - 2*cos(x)')
plt.grid(True)
plt.show()
```

```python
#e
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 5*sp.sin(2*x) + 3*x
integral = sp.integrate(f, (x, sp.pi, -sp.pi))

print("Valor da integral:", integral)

# Plotando o gráfico da função
x_vals = np.linspace(-np.pi, np.pi, 100)
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]

plt.plot(x_vals, f_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função 5*sin(2*x) + 3*x')
plt.grid(True)
plt.show()
```
