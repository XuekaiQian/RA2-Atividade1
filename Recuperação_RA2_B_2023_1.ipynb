# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

## 2.
$$g(x) = e^x \cos(x)$$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def ff(x):
    return 3*x**2 + 10*x - 7

def g(x):
    return np.exp(x) * np.cos(x)

def gg(x):
    return np.exp(x) * (np.cos(x) - np.sin(x))

x = np.linspace(-5, 5, 100)

plt.plot(x, f(x), label='f(x)')
plt.plot(x, ff(x), label="f_simplificada(x)")
plt.plot(x, g(x), label='g(x)')
plt.plot(x, gg(x), label="g_simplificada(x)")

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

# Questão 2
Resolva as derivadas a seguir explicitando todo o processo algébrico necessário para isso.

##1. 
Calcule a derivada da função: $g(x) = e^x \cos(x)$ para $x=5$.

##2. 
Calcule a derivada da função: $j(x) = \sqrt{x} \ln(x)$ para $x=3$


```python
import numpy as np
# Definir as funções e as derivadas
def g(x):
    return np.exp(x) * np.cos(x)

def gg(x):
    return np.exp(x) * (np.cos(x) - np.sin(x))

def j(x):
    return np.sqrt(x) * np.log(x)

def jj(x):
    return (2 + np.log(x)) / (2 * np.sqrt(x))
# Calcular os valores das derivadas no ponto específico que o questão pediu
gg_at_5 = gg(5)
jj_at_3 = jj(3)

print(f"O valor do g(x) é aproximadamente: {gg_at_5:.2f}")
print(f"O valor do j(x) é aproximadamente: {jj_at_3:.2f}")
```
